```python
import numpy as np
```

```python
x=np.linspace(0,20,50)
x
```

```python
plt.plot(x,x**2)
plt.plot(x,x**3)
plt.plot(x,x)
```

```python
help(plt.plot)
```

```python
plt.plot(x,x**2,color='c')
plt.plot(x,x**3,color='g')
plt.plot(x,x,color='r')
```

```python
plt.plot(x,x**2,color='c',linestyle='_')
plt.plot(x,x**3,color='g',linestyle='- -')
plt.plot(x,x,color='r',linestyle='_ ,')
```

```python
x2=np.linspace(10,30,50)
x2
```

```python
plt.plot(x2,np.sin(x2),color='yellow',marker='+',markersize=10)
plt.plot(x2,np.cos(x2),color='blue',marker='d',markersize=10,markerfacecolor='w')
```
